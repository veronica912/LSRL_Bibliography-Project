```python
import pandas as pd
import numpy as np
```

# Abstracts to be filled

```python
# open LSRL_Master2
worksheet = gc.open('LSRL_Master2').sheet1

# turn sheet into Pandas dataframe
df = pd.DataFrame(worksheet.get_all_records())
# replace empty abstracts with NaN
df.replace('', np.nan, inplace = True)
# drop entries with no title (so only LSRL 2022)
df.dropna(subset = ['Title'], axis = 0, inplace = True)

# turn title and abstract columns into dataframe
df_abstracts = df[['Title', 'Abstract']]

empty_abstracts = df_abstracts[df_abstracts['Abstract'].isna()]
empty_abstracts_title = empty_abstracts['Title'].to_list()
print('There are {} empty abstracts.'.format(len(empty_abstracts_title)))
```

```python
# find and mark all duplicate titles, including first and last occurrence
df_title_dup = df[df.duplicated(['Title'], keep = False)]
df_title_dup[['Title', 'Author']]
```
